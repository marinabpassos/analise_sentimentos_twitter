# Análise de Sentimentos com Twitter - Extração de Dados

Nesse notebook, iremos extrair dados da API do Twitter usando Tweepy. Faremos uma busca pelos mentions de 6 marcas. Salvaremos os Dataframes em arquivos a serem usados na análise

## Importação de libs

```python
import tweepy as tw
import pandas as pd
import time
```

## Autenticação para uso da API

Removi as chaves da API. Em caso de uso do código, utilizar chaves conseguidas pelo twitter developer

```python

API_KEY = 'substituir por seu valor'
SECRET_KEY = 'substituir por seu valor'
BEARER_TOKEN = 'substituir por seu valor'
ACCESS_TOKEN = 'substituir por seu valor'
SECRET_TOKEN = 'substituir por seu valor'
```

```python

auth = tw.OAuthHandler(API_KEY, SECRET_KEY)
auth.set_access_token(ACCESS_TOKEN, SECRET_TOKEN)
api = tw.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
```

## Busca de Tweets da Rico

```python
search_words = "@ricocomvc"
```

```python
counter = 0
tweet_dataset_rico = pd.DataFrame(columns=['Tweet Id', 'Tweet Date', 'Follower Count', 'Account Verified', 'Favorite Count', 'Retweets', 'Tweet Text'])

for tweet in tw.Cursor(api.search, tweet_mode='extended', q=search_words, lang="pt").items():
        if counter%100 == 0 and counter >0:
            print('Sleeping for 5 minutes')
            time.sleep(5*60)
            pass

        if counter%10==0:
            print('Tweet Downloaded: ', counter)


        appending_dataframe = pd.DataFrame([[tweet.id, tweet.created_at, tweet.user.followers_count, tweet.user.verified, tweet.favorite_count, tweet.retweet_count, tweet.full_text.encode('UTF-8')]], columns=['Tweet Id', 'Tweet Date', 'Follower Count', 'Account Verified', 'Favorite Count', 'Retweets', 'Tweet Text'])
        tweet_dataset_rico = tweet_dataset_rico.append(appending_dataframe)
        counter+=1
    

tweet_dataset_rico.to_excel('arquivos\tweet_dataset_rico.xlsx', index=False)
```
